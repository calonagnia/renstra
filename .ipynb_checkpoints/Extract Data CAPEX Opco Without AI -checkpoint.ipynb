```python
import os
import pandas as pd
import gspread
from google.auth.transport.requests import Request
from google.oauth2.credentials import Credentials
from google_auth_oauthlib.flow import InstalledAppFlow
from googleapiclient.discovery import build
from googleapiclient.errors import HttpError
import datetime  # Imported module to fix the datetime.datetime error

def connect_to_sheet():
    SCOPES = ['https://www.googleapis.com/auth/spreadsheets']
    creds = None
    if os.path.exists(r'C:\Users\mochamad.ilmawan\token.json'):
        creds = Credentials.from_authorized_user_file(r'C:\Users\mochamad.ilmawan\token.json', SCOPES)
    if not creds or not creds.valid:
        if creds and creds.expired and creds.refresh_token:
            creds.refresh(Request())
        else:
            flow = InstalledAppFlow.from_client_secrets_file(
                r'D:\OneDrive\Documents\Direct Selling\MONITORING\Monitoring 2023\client_secret_738271294618-6g8e0hle4jpq8nau1d7b3q9jfsgp0ruk.apps.googleusercontent.com.json', SCOPES)
            creds = flow.run_local_server(port=0)
        with open(r'C:\Users\mochamad.ilmawan\token.json', 'w') as token:
            token.write(creds.to_json())
    return creds

def update_sheet(spreadsheetId, spread_range, scopes, creds, df):
    df_clean = df.copy()
    for col in df_clean.columns:
        if pd.api.types.is_datetime64_any_dtype(df_clean[col]):
            df_clean[col] = df_clean[col].dt.strftime('%Y-%m-%d')
            
    df_clean = df_clean.astype(object).fillna('')
    header = df_clean.columns.tolist()
    data = df_clean.values.tolist()
    clean_data = []
    
    for row in data:
        clean_row = []
        for cell in row:
            # FIXED: Using proper datetime type checking
            if isinstance(cell, (datetime.datetime, pd.Timestamp)):
                clean_row.append(cell.strftime('%Y-%m-%d'))
            elif isinstance(cell, datetime.time):
                clean_row.append(cell.strftime('%H:%M:%S'))
            elif pd.isna(cell):
                clean_row.append('')
            else:
                clean_row.append(cell)
        clean_data.append(clean_row)
        
    values_with_header = [header] + clean_data
    try:
        service = build('sheets', 'v4', credentials=creds)
        service.spreadsheets().values().clear(spreadsheetId=spreadsheetId, range=spread_range, body={}).execute()
        body = {'values': values_with_header}
        result = service.spreadsheets().values().update(
            spreadsheetId=spreadsheetId, range=spread_range, valueInputOption="USER_ENTERED", body=body).execute()
        print(f"{result.get('updatedCells')} cells updated (sheet replaced).")
        return result
    except HttpError as error:
        print(f"An error occurred: {error}")
        return error

def find_cutoff_column(df, target_phrase="38. total spending / gr / sa"):
    for idx, col in enumerate(df.columns):
        if target_phrase.lower() in str(col).lower():
            return col, idx
    return None, None

def find_header_start_row(path, sheet_name, target_text="1 Tahun Capex"):
    df_raw = pd.read_excel(path, sheet_name=sheet_name, header=None, nrows=15)
    for idx, row in df_raw.iterrows():
        if row.astype(str).str.contains(target_text, case=False, na=False).any():
            return idx
    return None

def extract_data():
    files_and_sheets = [
        (r"C:\Users\mochamad.ilmawan\OneDrive - semenindonesia.com\Renstra\Capex\1. SP.xlsx", "SUMMARY"),
        (r"C:\Users\mochamad.ilmawan\OneDrive - semenindonesia.com\Renstra\Capex\2. ghopo rev.xlsx", "Prognose Spending"),
        (r"C:\Users\mochamad.ilmawan\OneDrive - semenindonesia.com\Renstra\Capex\3. SG .xlsx", "SUMMARY"),
        (r"C:\Users\mochamad.ilmawan\OneDrive - semenindonesia.com\Renstra\Capex\4. SMBR.xlsx", "SMBR 2026 R1"),
        (r"C:\Users\mochamad.ilmawan\OneDrive - semenindonesia.com\Renstra\Capex\5. ST.xlsx", "4000"),
        (r"C:\Users\mochamad.ilmawan\OneDrive - semenindonesia.com\Renstra\Capex\6. SBI.xlsx", "SUMMARY")
    ]

    all_dfs = []
    sg_columns = None
    
    for path, sheet in files_and_sheets:
        if os.path.exists(path):
            file_name = os.path.basename(path)
            print(f"Processing: {file_name} (Sheet: {sheet})...")
            
            # 1. DYNAMICALLY DETECT THE HEADER ROW
            header_start_idx = find_header_start_row(path, sheet)
            if header_start_idx is not None:
                # Assuming standard 3-level headers starting from discovered index
                chosen_header = [header_start_idx, header_start_idx + 1, header_start_idx + 2]
                print(f"   -> Found header starting at row index {header_start_idx}")
            else:
                chosen_header = [2, 3, 4]
                print("   -> '1 Tahun Capex' not found. Defaulting to header rows [2,3,4]")
                
            df = pd.read_excel(path, sheet_name=sheet, header=chosen_header)
            
            cleaned_cols = []
            for col in df.columns:
                parts = []
                # Ensure we handle multi-index tuple gracefully
                levels = col if isinstance(col, tuple) else [col]
                for level in levels:
                    level_str = str(level).strip().lower()
                    if "unnamed:" not in level_str and level_str != "nan" and level_str != "":
                        if not parts or parts[-1] != level_str:
                            parts.append(level_str)
                col_name = "_".join(parts) if parts else "unnamed"
                cleaned_cols.append(col_name)

            final_cols = []
            col_counts = {}
            for col in cleaned_cols:
                col_clean = col.replace('/', '_').replace('.', '')
                col_clean = '_'.join(filter(None, col_clean.split('_')))

                # 2. Use the CLEANED string for counting and appending
                if col_clean in col_counts:
                    col_counts[col_clean] += 1
                    final_cols.append(f"{col_clean}_{col_counts[col_clean]}")
                else:
                    col_counts[col_clean] = 0
                    final_cols.append(col_clean)

            # Assign headers before cutting/limiting dataframe
            df.columns = final_cols

            cutoff_col, cutoff_idx = find_cutoff_column(df)
            if cutoff_col:
                print(f"   -> Found target column '{cutoff_col}' at index {cutoff_idx}")
                df = df.iloc[:, :cutoff_idx + 1]
            else:
                print("   -> Target column not found. Keeping original dataframe structure.")
                df = df.copy()

            df = df.dropna(how='all')
            if "2 item capex" in df.columns:
                df = df[df["2 item capex"].notna() & (df["2 item capex"].astype(str).str.strip() != "")]
            if "15 % plan progres fisik (over all)" in df.columns:
                df["15 % plan progres fisik (over all)"] = pd.to_numeric(df["15 % plan progres fisik (over all)"], errors='coerce')
            df = df.ffill()
            df["Source_File"] = file_name
            
            # Capture the structure of sheet 3 ('3. SG .xlsx') to align structural variance
            if "3. SG" in file_name:
                sg_columns = list(df.columns)
                
            all_dfs.append(df)
        else:
            print(f"Error: File not found at {path}.")

    if all_dfs:
        combined_df = pd.concat(all_dfs, ignore_index=True)
        if sg_columns is not None:
            if "Source_File" not in sg_columns:
                sg_columns.append("Source_File")
            combined_df = combined_df.reindex(columns=sg_columns)
            print("\nSuccessfully aligned all data to the SG master framework!")
        else:
            print("\nWarning: '3. SG .xlsx' columns were not found. Data combined with default structure.")
        return combined_df
    return None

def main():
    combined_df = extract_data()
    if combined_df is not None:
        display(combined_df.head(10))
        print(f"Final Data Shape: {combined_df.shape[0]} rows x {combined_df.shape[1]} columns")
        creds = connect_to_sheet()
        update_sheet('1ckgGhsbbHGYfpqXtak_EG79ElY00bTvAADbNySRMa1s', 'Sheet1', None, creds, combined_df)

if __name__ == '__main__':
    main()
```
