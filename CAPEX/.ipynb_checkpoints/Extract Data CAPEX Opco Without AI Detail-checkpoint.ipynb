```python
import os
import re
import pandas as pd
import gspread
from google.auth.transport.requests import Request
from google.oauth2.credentials import Credentials
from google_auth_oauthlib.flow import InstalledAppFlow
from googleapiclient.discovery import build
from googleapiclient.errors import HttpError
import datetime  # Imported module to fix the datetime.datetime error

def connect_to_sheet():
    SCOPES = ['https://www.googleapis.com/auth/spreadsheets']
    creds = None
    if os.path.exists(r'C:\Users\mochamad.ilmawan\token.json'):
        creds = Credentials.from_authorized_user_file(r'C:\Users\mochamad.ilmawan\token.json', SCOPES)
    if not creds or not creds.valid:
        if creds and creds.expired and creds.refresh_token:
            creds.refresh(Request())
        else:
            flow = InstalledAppFlow.from_client_secrets_file(
                r'D:\OneDrive\Documents\Direct Selling\MONITORING\Monitoring 2023\client_secret_738271294618-6g8e0hle4jpq8nau1d7b3q9jfsgp0ruk.apps.googleusercontent.com.json', SCOPES)
            creds = flow.run_local_server(port=0)
        with open(r'C:\Users\mochamad.ilmawan\token.json', 'w') as token:
            token.write(creds.to_json())
    return creds

def update_sheet(spreadsheetId, spread_range, scopes, creds, df):
    df_clean = df.copy()
    for col in df_clean.columns:
        if pd.api.types.is_datetime64_any_dtype(df_clean[col]):
            df_clean[col] = df_clean[col].dt.strftime('%Y-%m-%d')
            
    df_clean = df_clean.astype(object).fillna('')
    header = df_clean.columns.tolist()
    data = df_clean.values.tolist()
    clean_data = []
    
    for row in data:
        clean_row = []
        for cell in row:
            # FIXED: Using proper datetime type checking
            if isinstance(cell, (datetime.datetime, pd.Timestamp)):
                clean_row.append(cell.strftime('%Y-%m-%d'))
            elif isinstance(cell, datetime.time):
                clean_row.append(cell.strftime('%H:%M:%S'))
            elif pd.isna(cell):
                clean_row.append('')
            else:
                clean_row.append(cell)
        clean_data.append(clean_row)
        
    values_with_header = [header] + clean_data
    try:
        service = build('sheets', 'v4', credentials=creds)
        service.spreadsheets().values().clear(spreadsheetId=spreadsheetId, range=spread_range, body={}).execute()
        body = {'values': values_with_header}
        result = service.spreadsheets().values().update(
            spreadsheetId=spreadsheetId, range=spread_range, valueInputOption="USER_ENTERED", body=body).execute()
        print(f"{result.get('updatedCells')} cells updated (sheet replaced).")
        return result
    except HttpError as error:
        print(f"An error occurred: {error}")
        return error

def find_cutoff_column(df, target_phrase="38 total spending  gr  sa"):
    for idx, col in enumerate(df.columns):
        if target_phrase.lower() in str(col).lower():
            return col, idx
    return None, None

def find_header_start_row(path, sheet_name, target_text="1 Tahun Capex"):
    df_raw = pd.read_excel(path, sheet_name=sheet_name, nrows=20, header=None)
    cleaned_target = target_text.lower().replace('.', '').replace(' ', '')
    for idx, row in df_raw.iterrows():
        cleaned_row = row.astype(str).str.lower().str.replace('.', '', regex=False).str.replace(' ', '', regex=False)
        if cleaned_row.str.contains(cleaned_target, regex=False, na=False).any():
            return idx
    return None

def extract_and_melt_all_capex():
    files_and_sheets = [
        (r"C:\Users\mochamad.ilmawan\OneDrive - semenindonesia.com\Renstra\Capex\2. Semen Padang - Laporan Capex s.d Juni 2026.xlsx", "TEMPLATE JULI 2026"),
        (r"C:\Users\mochamad.ilmawan\OneDrive - semenindonesia.com\Renstra\Capex\!Laporan Progres Capex 2026 GoPO 20260630 - kirim.xlsx", "Prognose Spending"),
        (r"C:\Users\mochamad.ilmawan\OneDrive - semenindonesia.com\Renstra\Capex\Laporan Progres Capex 2026 Opco PT Semen Gresik - Juli.xlsx", "SUMMARY"),
        (r"C:\Users\mochamad.ilmawan\OneDrive - semenindonesia.com\Renstra\Capex\3. SMBR - Rev.01 SMBR Laporan Spending Capex s.d 30 Juni 2026 (New Format).xlsx", "SMBR 2026 R1"),
        (r"C:\Users\mochamad.ilmawan\OneDrive - semenindonesia.com\Renstra\Capex\06 - (rev.) Kertas Kerja laporan spending capex 4000 (15 JUL 2026).xlsx", "NEW 4000"),
        (r"C:\Users\mochamad.ilmawan\OneDrive - semenindonesia.com\Renstra\Capex\6. SBI - SBB Summary Kertas Kerja Capex - Juni 2026.xlsx", "SUMMARY")
    ]

    months_2026 = [
        ("31/01/2026", "jan"), ("28/02/2026", "feb"), ("31/03/2026", "mar"),
        ("30/04/2026", "apr"), ("31/05/2026", "mei"), ("30/06/2026", "jun"),
        ("31/07/2026", "jul"), ("31/08/2026", "aug"), ("30/09/2026", "sep"),
        ("31/10/2026", "okt"), ("30/11/2026", "nov"), ("31/12/2026", "des")
    ]

    all_melted_rows = []

    for path, sheet in files_and_sheets:
        if not os.path.exists(path):
            print(f"Warning: File not found at {path}")
            continue
            
        file_name = os.path.basename(path)
        print(f"Processing & Reshaping: {file_name}...")
        
        # 1. Dynamically find header start row
        header_start_idx = find_header_start_row(path, sheet)
        chosen_header = [header_start_idx, header_start_idx + 1, header_start_idx + 2] if header_start_idx is not None else [2, 3, 4]
        
        df = pd.read_excel(path, sheet_name=sheet, header=chosen_header)
        df = df.dropna(how='all')
        
        cleaned_cols = []
        for col in df.columns:
            parts = []
            # Ensure we handle multi-index tuple gracefully
            levels = col if isinstance(col, tuple) else [col]
            for level in levels:
                level_str = str(level).strip().lower()
                if "unnamed:" not in level_str and level_str != "nan" and level_str != "":
                    if not parts or parts[-1] != level_str:
                        parts.append(level_str)
            col_name = "_".join(parts) if parts else "unnamed"
            cleaned_cols.append(col_name)

        final_cols = []
        col_counts = {}
        for col in cleaned_cols:
            # 1. Clean characters like slashes and dots
            col_clean = col.replace('/', '').replace('.', '')

            # 2. NEW: Remove leading numbers and trailing '_a'
            col_clean = re.sub(r'^\d+[\.\s]*', '', col_clean).strip()

            if col_clean.endswith('_a'):
                col_clean = col_clean[:-2]

            if col_clean in col_counts:
                col_counts[col_clean] += 1
                final_cols.append(f"{col_clean}_{col_counts[col_clean]}")
            else:
                col_counts[col_clean] = 0
                final_cols.append(col_clean)

        df.columns = final_cols

        # 2. Flatten Multi-index columns safely
        flattened_cols = []
        for col in df.columns:
            if isinstance(col, tuple):
                levels = [str(lvl).strip().lower() for lvl in col if pd.notna(lvl) and "unnamed:" not in str(lvl).lower()]
                flattened_cols.append("_".join(levels))
            else:
                flattened_cols.append(str(col).strip().lower())
        df.columns = flattened_cols
        
        id_col = None
        name_col = None
        
        # Look for ID column (Targeting "2. item capex" and "3. no project wbs")
        for col in df.columns:
            if any(k in col for k in ['item capex','2. item capex', '3_no_project_wbs', 'wbs', 'project', 'no_']):
                id_col = col
                break
                
        # Look for Name column
        for col in df.columns:
            if any(k in col for k in ['3. no. project / wbs', 'no. project  wbs', 'deskripsi', 'description', 'no project  wbs']) and col != id_col:
                name_col = col
                break
                
        if "entitas" in df.columns:
            df = df[df["entitas"].notna() & (df["entitas"].astype(str).str.strip() != "")]
        display(df.head(10))
        
        # FALLBACK: If keywords fail, assume column 0 is ID and column 1 is Name
        if id_col is None and len(df.columns) > 0:
            id_col = df.columns[0]
        if name_col is None and len(df.columns) > 1:
            name_col = df.columns[1]
            
        # Rename columns to standard schema
        df = df.rename(columns={id_col: 'item capex', name_col: 'Nama Capex'})
        df = df.dropna(subset=['item capex', 'Nama Capex'], how='all')

        # 4. Transpose monthly metric blocks
        for _, row in df.iterrows():
            proj_id = str(row.get('item capex', '')).strip()
            proj_name = str(row.get('Nama Capex', '')).strip()
            
            # Clean up default pandas string exports for missing fields
            if proj_id in ['nan', 'None'] or pd.isna(row.get('item capex')): proj_id = ''
            if proj_name in ['nan', 'None'] or pd.isna(row.get('Nama Capex')): proj_name = ''
            
            if proj_id == '' and proj_name == '':
                continue

            for date_str, month_keyword in months_2026:
                planning_val = 0
                realisasi_val = 0
                prognose_val = 0
                
                for raw_col, val in row.items():
                    col_lower = str(raw_col).lower()
                    if month_keyword in col_lower:
                        # Fallback to 0 if value is NaN/Null
                        clean_val = 0 if pd.isna(val) else val
                        
                        if 'plan' in col_lower:
                            planning_val = clean_val
                        elif 'real' in col_lower:
                            realisasi_val = clean_val
                        elif 'prog' in col_lower:
                            prognose_val = clean_val
                
                # Convert scalars to numeric objects cleanly
                all_melted_rows.append({
                    'item capex': proj_id,
                    'Nama Capex': proj_name,
                    'Month': date_str,
                    'Planning': float(pd.to_numeric(planning_val, errors='coerce') or 0),
                    'Realisasi': float(pd.to_numeric(realisasi_val, errors='coerce') or 0),
                    'Prognose': float(pd.to_numeric(prognose_val, errors='coerce') or 0),
                    'Source_File': file_name
                })

    final_df = pd.DataFrame(all_melted_rows)
    return final_df

def main():
    combined_monthly_df = extract_and_melt_all_capex()
    if combined_monthly_df is not None:
        display(combined_monthly_df.head(12))
        print(f"Final Data Shape: {combined_monthly_df.shape[0]} rows x {combined_monthly_df.shape[1]} columns")
        creds = connect_to_sheet()
        update_sheet('1ckgGhsbbHGYfpqXtak_EG79ElY00bTvAADbNySRMa1s', 'Detail Capex', None, creds, combined_monthly_df)

if __name__ == '__main__':
    main()
```
